```python
# Veri setini yeniden karıştırma
df = df.sample(frac=1)

# Özellikleri alma
features = df[['site_url', 'text_without_stopwords']]
features.head(5)

# site_url ve text_without_stopwords sütunlarını birleştirerek yeni bir sütun oluşturma
features['url_text'] = features["site_url"].astype(str) + " " + features["text_without_stopwords"]

#
```

```python
# Özellikler ve etiketlerin tanımlanması
X = features  # Özellikler
y = df['type']  # Etiketler

# Etiketleri bir Python listesine dönüştürme
y = y.tolist()
```

```python
# Veri kümesini ayırın ve TF-IDF kullanın
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

# Veri kümesini eğitim ve test kümelerine ayırın (veri, etiketler, test oranı ve rastgelelik durumu belirlenir)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# TF-IDF vektörize edicisini tanımlayın ve kullanın
tfidf_vectorizer = TfidfVectorizer(use_idf=True, stop_words='english')

# Eğitim verisi için TF-IDF dönüşümü yapın
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train['url_text'])

# Test verisi için TF-IDF dönüşümü yapın
X_test_tfidf = tfidf_vectorizer.transform(X_test['url_text'])
```

```python
# TF-IDF dönüşümü sonucu elde edilen X_train_tfidf verisini bir DataFrame'e dönüştürün
# ve sütun isimlerini TF-IDF vektörize edici ile oluşturulan özellik isimlerine ayarlayın
tfidf_train = pd.DataFrame(X_train_tfidf.A, columns=tfidf_vectorizer.get_feature_names_out())
```

```python
tfidf_train.head()  # Eğitim verisi TF-IDF dönüşümü sonrası ilk birkaç satırı görüntüler
```

```python
from sklearn import metrics

# AdaBoost sınıflandırıcısını oluşturun, karar ağacı sınıflandırıcısıyla (max_depth=10) birlikte ve 5 stimatör kullanarak
Adab = AdaBoostClassifier(DecisionTreeClassifier(max_depth=10), n_estimators=5, random_state=1)

# Eğitim verisi üzerinde AdaBoost sınıflandırıcısını eğitin
Adab.fit(tfidf_train, y_train)

# Test verisi üzerinde tahmin yapın
y_pred3 = Adab.predict(X_test_tfidf)

# Doğruluk skorunu hesaplayın
ABscore = metrics.accuracy_score(y_test, y_pred3)
print("accuracy: %0.3f" %ABscore) # Doğruluk skorunu yazdırın
```
